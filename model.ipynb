#REMOVING THE USELESS COLUMNS

```python
df = df[["Contract", "Tenure Months", "Monthly Charges", "Tech Support", "Internet Service", "Payment Method", "Churn Label"]]
df.columns
```

```python
df = pd.get_dummies(
    df,
    columns=["Contract", "Tech Support", "Internet Service", "Payment Method", "Churn Label"],
    drop_first=True
)
```

```python
df.head()
```

```python
df.columns
```

```python
df.columns = df.columns.str.replace(" ", "_")
df.columns = df.columns.str.replace(r"[()]", "", regex=True)
```

```python
df.columns
```

```python
df.rename(columns={"Churn_Label_Yes": "Churn"}, inplace=True)
```

```python
df.columns
```

MODELS CREATION AND EVALUATIONS

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop("Churn", axis=1)
Y = df["Churn"]

X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.2, random_state=42
)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
Y_train.shape
```

```python
Y_test.shape
```

```python
# 1. Logistic Regression
from sklearn.linear_model import LogisticRegression
log_model = LogisticRegression(class_weight='balanced')
log_model.fit(X_train, Y_train)

# 2. Random Forest
from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier()
rf_model.fit(X_train, Y_train)

# 3. XGBoost
from xgboost import XGBClassifier
xgb_model = XGBClassifier(eval_metric='logloss')
xgb_model.fit(X_train, Y_train)
```

```python
# Predictions
probs = log_model.predict_proba(X_test)[:, 1]
custom_pred = (probs > 0.3).astype(int)
rf_pred = rf_model.predict(X_test)
xgb_pred = xgb_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report

print("Logistic Regression\n")
print("Accuracy:", accuracy_score(Y_test, custom_pred))
print(classification_report(Y_test, custom_pred))

print("Random Forest\n")
print("Accuracy:", accuracy_score(Y_test, rf_pred))
print(classification_report(Y_test, rf_pred))

print("XGBoost\n")
print("Accuracy:", accuracy_score(Y_test, xgb_pred))
print(classification_report(Y_test, xgb_pred))
```
